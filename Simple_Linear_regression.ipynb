```python
#PANDAS is consider as Class/library used for Data
import pandas as py 
#py -> variable
#RUN (shortcutkey) -> Shift enter 
```

```python
dataset=py.read_csv("Salary_data.csv") #make sure file under coresponding folder
# dataset-> variable,py call to read the file 
# read_csv -> pandas function
```

```python
dataset
#debug to check the value i.e what is available in the file
#years of experience -> Input , Salary -> Output
```

```python
input_value=dataset[["YearsExperience"]]
#input_value-> variable, want to get first column as input so i pass column name with []
```

```python
input_value #view the data
```

```python
output_value=dataset[["Salary"]]
#output_value-> variable, want to get second column as output so i pass column name with []
```

```python
output_value #view the value
```

```python
import matplotlib.pyplot as plt #class matplotlib call function pyplot to assign var ->plt
```

```python
#pass input and output value
plt.scatter(input_value,output_value)
#label x and y to display and show
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
#sklearn->library used for machinelearning
# this libray call model_selection class use only train_test_split function to perform "train set" ->memory save
```

```python
train_test_split #call to value
```

```python
train_test_split(input_value,output_value,test_size=0.30,random_state=0)
#call the train set of ip/op/size/state
#test_size= 30 (70-> training set:30 test set)
```

```python
#multiple value coming, so we assigh the valid position (x ->train input , x->test input , y->train output, y->test output)
# X ->input Y -> Output
# -> Train Set
# -> Test Set, 
```

```python
X_ipval_train,X_iptestset,y_opval_train,y_optestset = train_test_split(input_value,output_value,test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression #Algoritm (for creating Model)
regressor=LinearRegression() # variable -> (regressor) to call the function LinearRegression()
#above 2 is procedure
```

```python
#regressor.fit(X_ipval_train,y_opval_train) #pass the value -> weight and bias calculate -> y=mx+b

regressor.fit(X_ipval_train,y_opval_train)
```

```python
weight=regressor.coef_ #coefficient variable, weight->SLOPE
weight
```

```python
bais=regressor.intercept_ #intercept variable, bias->start point
bais
```

```python
#Evalution metrics
y_pred=regressor.predict(X_iptestset)
```

```python
#Evalution metrics
from sklearn.metrics import r2_score
r_score=r2_score(y_optestset,y_pred)
```

```python
r_score
```

```python
#nearly to 1-> so model perform well
#nearly to 0 -> Bad Model
```

```python
#save the model import the pickle library
import pickle
filename="Finalized_Linear_Model.sav"
```

```python
pickle.dump(regressor,open(filename,'wb')) #wb->write binary
```

```python
#check the model before deployment
loaded_model=pickle.load(open("Finalized_Linear_Model.sav",'rb')) #Read binary
result=loaded_model.predict([[13]])
```
